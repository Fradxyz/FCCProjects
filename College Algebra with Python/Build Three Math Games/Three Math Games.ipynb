**Scatter Plot Game**

```python
import matplotlib.pyplot as plt
import numpy as np
import random
import re

# Random zoom
zooms = [5, 10, 20, 50, 100]
ticks = [1, 2, 5, 10, 20]

zoom_index = random.randint(0, len(zooms)-1)
zoom = zooms[zoom_index]
tick = ticks[zoom_index]

# Set up graph
x_min = y_min = -zoom
x_max = y_max = zoom

fig, ax = plt.subplots()
plt.axis([x_min, x_max, y_min, y_max])
plt.plot([x_min, x_max], [0, 0], 'k')
plt.plot([0, 0], [y_min, y_max], 'k')

ax.set_title('Scatter Plot Game')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xticks(np.arange(x_min, x_max+tick, tick))
ax.set_yticks(np.arange(y_min, y_max+tick, tick))
ax.grid(True)

# Plot random point
rand_x = random.randint(x_min, x_max)
rand_y = random.randint(y_min, y_max)
plt.plot(rand_x, rand_y, 'ro')

plt.show()

# Guess
user_input = input("\nGuess the coordinate of the point (x, y): ")
result = re.search(r"^\((.+),\s*(.+)\)$", user_input)

user_x = int(result.group(1))
user_y = int(result.group(2))

if user_x == rand_x and user_y == rand_y:
  print(f"That's correct! The coordinate of the point is ({rand_x}, {rand_y})")
else:
  print(f"That's incorrect. The coordinate of the point is actually ({rand_x}, {rand_y})")
```

**Algebra Practice Game**

```python
import random
from sympy import *

print("Solve the following problem: ")

# Generate problem
step_range = random.randint(0, 1) + 2  # one-step or two-step
int_lst = []
operations_lst = []

# random integers
for i in range(step_range):
  int_lst.append(random.randint(-10, 10))

# random operations
for i in range(len(int_lst)-1):
  operations = ['*', '/', '+', '-']
  operations_lst.append(operations[random.randint(0, len(operations)-1)])

eq_str  = ""

for i, n in enumerate(int_lst):
  if n < 0:
    n = f"({n})"
  if i == len(int_lst)-1:
    eq_str += f"{n}"
  else:
    eq_str += f"{n} {operations_lst[i]} "


eq = eq_str + " - x"
print(f"x = {eq_str}")

# Solve problem
x = symbols('x')
solution = round(float(solve(eq, x)[0]), 2)

user_solution = round(float(input("x = ")), 2)

if solution == user_solution:
  print(f"That's correct! The solution is {solution}")
else:
  print(f"Try again. The correct solution is {solution}")
```
