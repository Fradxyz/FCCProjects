Print the column names and first two rows

```python
table_1 = pd.read_csv('https://people.sc.fsu.edu/~jburkardt/data/csv/faithful.csv')
table_1.head(2)
```

Store the column names as a list

```python
column_names = table_1.columns

for col in column_names:
  print(col)
```

Convert data to numpy arrays

```python
col_name1 = column_names[0]
col_name2 = column_names[1]

x = table_1[col_name1].to_numpy()
y = table_1[col_name2].to_numpy()

print(f"{col_name1}: {x}")
print(f"{col_name2}: {y}")
```

Graph data

```python
x_min = x.min()
x_max = x.max()
y_min = 0
y_max = 6

fig, ax = plt.subplots()
plt.axis([x_min, x_max, y_min, y_max])

ax.set_xlabel(col_name1)
ax.set_ylabel(col_name2)

plt.plot(x, y, 'go')
plt.show()
```
